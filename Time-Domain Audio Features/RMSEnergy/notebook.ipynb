```python
voice1, sr1 = librosa.load(voice1_file)
voice2, sr2 = librosa.load(voice2_file)
voice3, sr3 = librosa.load(voice3_file)

print(f'corresponding number of samples are: {voice1.size}, {voice2.size}, {voice3.size}')
```

```python

plt.figure(figsize=(16,15))

plt.subplot(3,1,1)
librosa.display.waveshow(voice1, alpha=0.7)
plt.title('Debussy waveform')
plt.ylim(-1,1)

plt.subplot(3,1,2)
librosa.display.waveshow(voice2, alpha=0.7)
plt.title('Duke waveform')
plt.ylim(-1,1)

plt.subplot(3,1,3)
librosa.display.waveshow(voice3, alpha = 0.7)
plt.title('redhot waveform')
plt.ylim(-1,1)
```

```python
FRAME_SIZE = 1024
HOP_LENGTH = 512
def rmsenergy(signal):
    return np.array([np.sqrt(sum((signal[i:i+FRAME_SIZE]**2)/FRAME_SIZE)) for i in range(0, len(signal), HOP_LENGTH)])
```

```python
debussy_rms = rmsenergy(voice1)
duke_rms = rmsenergy(voice2)
redhot_rms = rmsenergy(voice3)
```

```python
plt.figure(figsize=(10,15))
frames = range(0, debussy_rms.size)
time = librosa.frames_to_time(frames,hop_length=HOP_LENGTH)

plt.subplot(3,1,1)
librosa.display.waveshow(voice1, alpha=0.7)
plt.plot(time, debussy_rms, color='red')
plt.title('Voice1 waveform')
plt.ylim(-.25,.25)
plt.xlim(2,3)

plt.subplot(3,1,2)
librosa.display.waveshow(voice2, alpha=0.7)
plt.plot(time, duke_rms, color='red')
plt.title('Voice2 waveform')
plt.ylim(-.25,.25)
plt.xlim(2,3)

plt.subplot(3,1,3)
librosa.display.waveshow(voice3, alpha = 0.7)
plt.plot(time, redhot_rms, color='red')
plt.title('Voice3 waveform')
plt.ylim(-.25,.25)
plt.xlim(2,3)

'''
Using librosa rms function:
rms_debussy = librosa.feature.rms(debussy, frame_length=FRAME_SIZE, hop_length=HOP_LENGTH)[0] # to reduce dimension
rms_redhot = librosa.feature.rms(redhot, frame_length=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
rms_duke = librosa.feature.rms(duke, frame_length=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
'''
```
